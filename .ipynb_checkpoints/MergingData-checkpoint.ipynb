```python
opinion = pd.read_csv("YCOM_2020_Data.csv")
# Source US Bureau of Economic Analysis
income = pd.read_csv("income_data_cleaned.csv")
CoalProductionbyCountyandState = pd.read_csv("CoalProductionbyCountyandState.csv")
CoalProductionbyCountyandState = CoalProductionbyCountyandState.rename(columns={'Coal-Producing State and County' : 'GeoName'})

opinion[opinion.GeoName]
```

```python
sep = ','
opinion['GeoName'] = opinion['GeoName'].apply(lambda x: x.split(sep, 1)[0])
opinion
```

```python
income = income.rename(columns={'Location' : 'GeoName'})
income
```

```python
coal_opinion = pd.merge(opinion, CoalProductionbyCountyandState, on='GeoName', how='outer')
states = ["Alaska", "Alabama", "Arkansas", "American Samoa", "Arizona", "California", "Colorado", "Connecticut", "District of Columbia", "Delaware", "Florida", "Georgia", "Guam", "Hawaii", "Iowa", "Idaho", "Illinois", "Indiana", "Kansas", "Kentucky", "Louisiana", "Massachusetts", "Maryland", "Maine", "Michigan", "Minnesota", "Missouri", "Mississippi", "Montana", "North Carolina", "North Dakota", "Nebraska", "New Hampshire", "New Jersey", "New Mexico", "Nevada", "New York", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Puerto Rico", "Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Virginia", "Virgin Islands", "Vermont", "Washington", "Wisconsin", "West Virginia", "Wyoming"]
coal_opinion.drop_duplicates(subset='GeoName', keep='first', inplace=True)
coal_opinion = coal_opinion.dropna()
coal_opinion
```

```python
sns.scatterplot(x=coal_opinion.CO2limits[23:], y=coal_opinion['Total Production'][23:])
```
