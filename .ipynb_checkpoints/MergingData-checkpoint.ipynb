```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stat
import numpy as np
import scipy.stats as st
```

```python
opinion = pd.read_csv("YCOM_2020_Data.csv")
# Source US Bureau of Economic Analysis
income = pd.read_csv("income_data_cleaned.csv")
CoalProductionbyCountyandState = pd.read_csv("CoalProductionbyCountyandState.csv")
CoalProductionbyCountyandState = CoalProductionbyCountyandState.rename(columns={'Coal-Producing State and County' : 'GeoName'})
```

```python
opinion.columns
```

```python
sep = ','
opinion['GeoName'] = opinion['GeoName'].apply(lambda x: x.split(sep, 1)[0])
opinion
```

```python
income = income.rename(columns={'Location' : 'GeoName'})
income
```

```python
coal_opinion = pd.merge(opinion, CoalProductionbyCountyandState, on='GeoName', how='outer')
states = ["Alaska", "Alabama", "Arkansas", "American Samoa", "Arizona", "California", "Colorado", "Connecticut", "District of Columbia", "Delaware", "Florida", "Georgia", "Guam", "Hawaii", "Iowa", "Idaho", "Illinois", "Indiana", "Kansas", "Kentucky", "Louisiana", "Massachusetts", "Maryland", "Maine", "Michigan", "Minnesota", "Missouri", "Mississippi", "Montana", "North Carolina", "North Dakota", "Nebraska", "New Hampshire", "New Jersey", "New Mexico", "Nevada", "New York", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Puerto Rico", "Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Virginia", "Virgin Islands", "Vermont", "Washington", "Wisconsin", "West Virginia", "Wyoming"]
coal_opinion.drop_duplicates(subset='GeoName', keep='first', inplace=True)
coal_opinion = coal_opinion.dropna()
coal_opinion = coal_opinion[coal_opinion['GeoType'] != 'State']
coal_opinion
```

```python
sns.scatterplot(x=coal_opinion.CO2limits, y=coal_opinion['Total Production'])
```

```python
coal_opinion = coal_opinion[coal_opinion['GeoType'] != 'State']
coal_countries = coal_opinion['GeoName'].to_list()
coal_countries
coal_opinion_ID = coal_opinion[coal_opinion['GeoType'] != 'State']
coal_countries_ID = coal_opinion['GEOID'].to_list()
coal_countries_ID
```

```python
opinion['coal_county'] = opinion.GEOID.isin(coal_countries_ID)
opinion[opinion['coal_county'] == True]
opinion.columns
```

```python
opinion_sample = opinion[opinion['GeoType'] != 'State'].sample(90)
opinion_coalOnly = opinion[opinion['coal_county'] == True]
coal_opinion_sample = pd.concat([opinion_coalOnly, opinion_sample]).drop_duplicates()
sns.scatterplot(x='human', y='humanOppose', hue='coal_county', data=coal_opinion_sample)
```

```python
sns.set_theme(style="whitegrid")
# coal_opinion_barplot = []
# coal_opinion_barplot.append([stat.mean(opinion[opinion['coal_county'] == True]['human']), 
#                              stat.stdev(opinion[opinion['coal_county'] != True]['human']),
#                              True])
# coal_opinion_barplot.append([stat.mean(opinion[opinion['coal_county'] != True]['human']),
#                              stat.stdev(opinion[opinion['coal_county'] == True]['human']),
#                             False])
# coal_opinion_barplot = pd.DataFrame(coal_opinion_barplot, columns=['mean', 'stddev', 'coal_county'])
# coal_opinion_barplot
topic_data = coal_opinion_sample
topic = 'fundrenewables'
print(st.t.interval(alpha=0.90, df=len(topic_data[topic_data['coal_county'] == True][topic])-1,
              loc=np.mean(topic_data[topic_data['coal_county'] == True][topic]),
              scale=st.sem(topic_data[topic_data['coal_county'] == True][topic])))
print(st.t.interval(alpha=0.90, df=len(topic_data[topic_data['coal_county'] != True][topic])-1,
              loc=np.mean(topic_data[topic_data['coal_county'] != True][topic]),
              scale=st.sem(topic_data[topic_data['coal_county'] != True][topic])))
g = sns.barplot(
    data=coal_opinion_sample,
    x='coal_county', y='human', hue='coal_county',
    ci=90, palette="dark", alpha=.6, 
)
# WE SHOULD COMPARE TOP renewable counties to coal counties
# maybe we could do regression with pop, income, and race data but idk meh
# statistical diff appear for 'human', 'president', haven't checked all :()
```
