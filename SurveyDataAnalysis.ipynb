```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("YCOM_2020_Data.csv")
df.head()
```

```python

# make data:
np.random.seed(3)
x = df['GeoName'].to_numpy()[1:50]
y = df['TotalPop'].to_numpy()[1:50]

```

```python
# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

# ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
#        ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
